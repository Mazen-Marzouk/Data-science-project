# **Invistigate File**

```python
match_data = pd.read_csv("Match.csv")
# this code read file ("Match.csv") and save in path could (match_data)
```

#**Define Function**

```python
def barplot(x, y, data=match_data):
    sns.barplot(x=x , y=y , data=match_data)
    plt.title(f"Comparision between {x} and {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
#Define Function to make barplot
```

# ***Cleaning***

```python
match_data_clened = match_data.fillna(match_data.median) #change nan values to values depending on median
```

```python
match_data_clened.isna().sum() #check that there isn't any nun values
```

```python
match_data_clened.info() #check that information of all columns are complete
```

Now we remove all nun values and replace it to integers depending on median

# **EDA**

```python
sns.barplot(x="match_api_id" ,hue="league_id", data=match_data_clened)
plt.title("Comparision between Leagues")
plt.xlabel("match_api_id")
plt.ylabel("count")
plt.show()
#make barplot to show comparison between Leagues
```

Here we find that Portugal League (ZON Sarges) is the most league in match_api

```python
barplot('home_team_goal','league_id',data=match_data)
#make barplot to show the most league in home_team_goal
```

```python
sns.scatterplot(x=match_data_clened["stage"],y=match_data_clened["country_id"],hue=match_data_clened["league_id"])
plt.show()
#make scattershot to show comparison between countries and stages of Leagues
```

We find that alot of Leagues are in the same stage as Ligue1(France),Premier League(England),SerieA(Italy),etc..
